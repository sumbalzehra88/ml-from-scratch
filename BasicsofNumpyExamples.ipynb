Example 8: Boolean Indexing

```python
names = np.array(['Bob', 'Joe', 'Will', 'Bob', 'Will', 'Joe',
'Joe'])
data = np.random.randn(7, 4)
names
```

```python
data
```

```python
names == 'Bob'
```

```python
data[names == 'Bob']
```

```python
data[names == 'Bob', 2:]
```

```python
data[names == 'Bob', 3]
```

```python
names != 'Bob'
```

```python
data[~(names == 'Bob')]
```

```python
mask = (names == 'Bob') | (names == 'Will')
mask
```

```python
data[mask]
```

```python
data[data < 0] = 0
data
```

```python
data[names != 'Joe'] = 7
data
```
