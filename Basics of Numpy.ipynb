Broadcasting

Because of broadcasting we are able to perform vectorization operation.

3x3
3x1

if value is same or the value is 1 then broadcasting is possible

5x4
  4

if the size is same and where the dimensions are not present, it adds 1 automatically.
so in this case 4 will be changed to 1x4 and it will be compatible.

256x256x64
        64

in the case of a 3d array (shape of an array), if a 1d array has to be applied on it like in this case, then it can be broadcasted in this case.
