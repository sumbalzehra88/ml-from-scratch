Question 12

```python
import numpy as np

marks = np.array([
    [60, 70, 50],   # Subject 1
    [40, 60, 80],   # Subject 2
    [70, 55, 65]    # Subject 3
])

credit_hours = np.array([2, 3, 4])
```

```python
grade_points = np.where(marks >= 80, 4.0,
                np.where(marks >= 75, 3.4,
                np.where(marks >= 70, 3.0,
                np.where(marks >= 67, 2.7,
                np.where(marks >= 64, 2.4,
                np.where(marks >= 60, 2.0,
                np.where(marks >= 57, 1.7,
                np.where(marks >= 54, 1.4,
                np.where(marks >= 50, 1.0, 0.0)))))))))
```

```python
weighted_gp = grade_points * credit_hours[:, np.newaxis]
```

```python
total_credits = credit_hours.sum()
cgpa = weighted_gp.sum(axis=0) / total_credits
cgpa
```

Question 14

```python
import numpy as np

A = np.random.rand(4, 4)
print("Matrix A:\n", A)

dot_product = np.dot(A, A)
print("\nDot Product (A·A):\n", dot_product)

inverse_A = np.linalg.inv(A)
print("\nInverse of A:\n", inverse_A)
```

Question 15

```python
import numpy as np

A = np.random.rand(4, 4)

dot_product = np.dot(A, A)

inverse_A = np.linalg.inv(A)

np.save("matrix_A.npy", A)
np.save("dot_product.npy", dot_product)
np.save("inverse_A.npy", inverse_A)

np.savez("all_arrays.npz", A=A, dot_product=dot_product, inverse_A=inverse_A)

print("Arrays saved to disk!")
```
