```python
#applying one hot encoding on Pitch_Type
#applying one hot encoding on Stage
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

# Initialize OneHotEncoder
# sparse_output=False gives you a regular array (not sparse matrix)
# drop='first' prevents multicollinearity (optional)
encoder = OneHotEncoder(sparse_output=False, drop=None)

# Fit and transform Stage
pitch_encoded = encoder.fit_transform(cleaned_data[['Pitch_Type']])

# Get feature names
feature_names = encoder.get_feature_names_out(['Pitch_Type'])

# Create DataFrame with encoded columns
pitch_df = pd.DataFrame(pitch_encoded, columns=feature_names, index=cleaned_data.index)

# Drop original column and add encoded columns
cleaned_data = cleaned_data.drop('Pitch_Type', axis=1)
cleaned_data = pd.concat([cleaned_data, pitch_df], axis=1)

print("✓ Applied OneHotEncoder to Pitch_Type")
print(f"Created columns: {list(feature_names)}")
cleaned_data.head()
```

```python
#applying one hot encoding on Stage
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

# Initialize OneHotEncoder
# sparse_output=False gives you a regular array (not sparse matrix)
# drop='first' prevents multicollinearity (optional)
encoder = OneHotEncoder(sparse_output=False, drop=None)

# Fit and transform Stage
pitch_encoded = encoder.fit_transform(cleaned_data[['Stage']])

# Get feature names
feature_names = encoder.get_feature_names_out(['Stage'])

# Create DataFrame with encoded columns
pitch_df = pd.DataFrame(pitch_encoded, columns=feature_names, index=cleaned_data.index)

# Drop original column and add encoded columns
cleaned_data = cleaned_data.drop('Stage', axis=1)
cleaned_data = pd.concat([cleaned_data, pitch_df], axis=1)

print("✓ Applied OneHotEncoder to Stage")
print(f"Created columns: {list(feature_names)}")
cleaned_data.head()
```

```python
#applying binary encoding on Toss_Decision

cleaned_data['Toss_Decision'].replace({'Bat' : 1, 'Field' : 0}, inplace=True)   
cleaned_data.head() 
```

```python
#applying binary encoding to toss_winner

cleaned_data['Toss_Winner'].replace({'Team_A' : 1, 'Team_B' : 0}, inplace=True)   
cleaned_data.head() 
```

```python
#applying binary encoding to winner

cleaned_data['Winner_y'].replace({'Team_A' : 1, 'Team_B' : 0}, inplace=True)   
cleaned_data.head() 
```

```python
# Creating new features from existing features

cleaned_data['Form_Difference'] = cleaned_data['Team_A_Form'] - cleaned_data['Team_B_Form']
cleaned_data['Tech_Index_Difference'] = cleaned_data['Team_A_Tech_Index'] - cleaned_data['Team_B_Tech_Index']
cleaned_data.head()
```

```python
cleaned_data.drop(columns=['Team_A_Form','Team_B_Form','Team_A_Tech_Index','Team_B_Tech_Index'], inplace=True)
cleaned_data.head()
```

Features like pitch_type that have more than two categories are encoded using one-hot encoding, while binary features like toss_decision and toss_winner are encoded using binary encoding. This ensures that the categorical data is transformed into a numerical format suitable for machine learning algorithms.

Furthermore, it makes it easy to analyze the distribution of different categories within these features, which can provide insights into their potential impact on match outcomes.

The features like form and teach index of each team are used to find the difference of form and tech index among the two teams. This helps in understanding better that which team is performing better.

If the values are negative then Team B is the winner. If the values are inclined towards the postive, then Team A is the winner
